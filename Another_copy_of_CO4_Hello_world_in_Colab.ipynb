````python
import pandas as pd
from datetime import datetime

# Example data structure (you could also load from CSV or Excel)
data = {
    "Item": [
        "car battery","groceries","tissue","lawn equipment","boost drink",
        "oil change","wired mouse","gasoline","lottery","rock salt"
    ],
    "Total Cost":[311.01,567.76,22.25,438.82,640.77,102.17,14.07,113.15,921.30,85.00],
    "Purchase Date":[
        "2026-02-03","2026-03-28","2025-08-23","2025-09-04","2025-11-13",
        "2026-02-26","2026-02-22","2026-01-14","2024-11-25","2025-11-25"
    ],
    "Target Daily Cost":[3.25,15.00,0.06,1.75,3.75,1.65,0.20,1.00,1.50,0.50]
}

df = pd.DataFrame(data)

# Convert purchase date
df["Purchase Date"] = pd.to_datetime(df["Purchase Date"])

# Today's date
today = pd.Timestamp.today().normalize()

# Days since purchase
df["Days Since Purchase"] = (today - df["Purchase Date"]).dt.days

# Cost per day
df["Cost Per Day"] = df["Total Cost"] / df["Days Since Purchase"]

# Break-even days
df["Break-even Days"] = df["Total Cost"] / df["Target Daily Cost"]

# Break-even date
df["Break-even Date"] = df["Purchase Date"] + pd.to_timedelta(df["Break-even Days"], unit="D")

# Remaining days
df["Remaining Days"] = (df["Break-even Date"] - today).dt.days

# Percent completed
df["Percent Completed"] = (df["Days Since Purchase"] / df["Break-even Days"]) * 100

# Monthly equivalent
df["Per Month"] = df["Cost Per Day"] * 30

# Round values
df = df.round({
    "Cost Per Day":2,
    "Break-even Days":2,
    "Percent Completed":2,
    "Per Month":2
})

print(df)
````

````python
def show_summary(df):

    if len(df) == 0:
        print("\nNo purchases recorded.")
        return

    # Use format='mixed' to handle potentially inconsistent date formats
    # Use errors='coerce' to turn unparseable dates into NaT, ensuring datetime dtype
    df["Purchase Date"] = pd.to_datetime(df["Purchase Date"], format='mixed', errors='coerce')

    today = datetime.today().date()

    df["Days Since Purchase"] = (today - df["Purchase Date"].dt.date).dt.days
    df["Cost Per Day"] = df["Total Cost"] / df["Days Since Purchase"].replace(0,1)

    total_spent = df["Total Cost"].sum()
    total_items = len(df)

    most_expensive = df.loc[df["Total Cost"].idxmax()]

    avg_cost_day = df["Cost Per Day"].mean()

    print("\n========== SUMMARY ==========")
    print("Total Items:", total_items)
    print("Total Money Spent: $", round(total_spent,2))
    print("Average Cost Per Day: $", round(avg_cost_day,2))
    print("Most Expensive Item:", most_expensive["Item"],
          "($"+str(round(most_expensive["Total Cost"],2))+")")
    print("=============================\n")
````

````python
def apply(criteria, n):
    count = 0
    for i in range(n+1):
         if criteria(i):
            count += 1
    return count


# What this function does:

# criteria -> a function that tests a condition (like “is it even?”)

# n -> the upper limit of numbers to test

# How it works:

# count = 0
# Starts a counter at zero.

# for i in range(n + 1)
# Loops through numbers from 0 to n, including n.

# if criteria(i):
# Calls the criteria function on each number.
# If it returns True, then…

# count += 1
# …increase the counter.

# return count
# After the loop finishes, return how many numbers satisfied the criteria.


def is_even(x):
    return x % 2 == 0

# This returns True if x is divisible by 2, otherwise False.

# So it is used as a test for “Is this number even?”


print(apply(is_even, 10))

# This asks:

# “How many numbers from 0 to 10 are even?”

# Even numbers between 0 and 10:

# 0, 2, 4, 6, 8, 10 -> 6 numbers

# Final Output
# 6
````

````python
class Question:
    def __init__(self, prompt, answer):
        self.prompt = prompt
        self.answer = answer


question_prompts = ["What color are apples?\n(a) Red/Green\n(b) Purple\n(c) Orange\n\n",
 "What color are bananas?\n(a) Teal\n(b) Magenta\n(c) Yellow\n\n",
 "What color are strawberries?\n(a) Yellow\n(b) Red\n(c) Blue\n\n"]

questions = [Question(question_prompts[0], "a"),
             Question(question_prompts[1], "c"),
             Question(question_prompts[2], "b"),]

def run_test(questions):
    score = 0
    for question in questions:
        answer = input(question.prompt)
        if answer == question.answer:
            score += 1
    print("You got " + str(score) + "/" + str(len(questions)) + " correct")

run_test(questions)
````

````python
import random

lowest_number = 26
highest_number = 100
answer = random.randint(lowest_number, highest_number)
guesses = 0

print("Python Number Guessing Game")
print(f"Select a number between {lowest_number} and {highest_number}")

while True:
    guess = input("Enter your guess" )

    if guess.lower() == "q":
        print("Game exited.")
        break

    if guess.isdigit():
        guess = int(guess)
        guesses += 1

        if guess < lowest_number or guess > highest_number:
            print("That number is out of range")
        elif guess < answer:
            print("Too low! Try again!")
        elif guess > answer:
            print("Too high! Try again!")
        else:
            print(f"CORRECT! The answer was {answer}")
            print(f"Number of guesses: {guesses}")
            break
    else:
        print("Invalid guess. Please enter a number.")

````

````python
from os.path import basename
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  def __init__(self, radius):
    self.radius = radius

  def area(self):
    return 3.14 * self.radius ** 2

class Square(Shape):
  def __init__(self, side):
    self.side = side

  def area(self):
    return self.side ** 2

class Triangle(Shape):
  def __init__(self, base, height):
    self.base = base
    self.height = height

  def area(self):
    return self.base * self.height * 0.5

shapes = [Circle(4), Square(5), Triangle(6, 7)]

for shape in shapes:
  print(shape.area())
````

````python
class Shape:
  def __init__(self, color, is_filled):
    self.color = color
    self.is_filled = is_filled


class Circle(Shape): # Renamed from Circles to match object instantiation
  def __init__(self, color, is_filled, radius):
    super().__init__(color, is_filled)

    self.radius = radius

class Square(Shape):
  def __init__(self, color, is_filled, width): # Added __init__ method
    super().__init__(color, is_filled)
    self.width = width


class Triangle(Shape):
  def __init__(self, color, is_filled, width, height):
    super().__init__(color, is_filled)
    self.width = width
    self.height = height

circle = Circle(color="red", is_filled=True, radius=5)
print(circle.color)
print(circle.is_filled)
print(circle.radius)
````

````python

def char_counts(s):
    (c, v) = (0, 0)
    vowels = "aeiou"
    for char in s:
      if char in vowels:
        v += 1
      else:
        c += 1

    return (v, c)
print(char_counts("abcd"))
print(char_counts("zcght"))
print(char_counts("aeiou"))
print(char_counts("dklox"))
````

````python
def f(expr, old_list, test = lambda x: True):
    new_list = []
    for item in old_list:
        if test(item):
            new_list.append(expr(item))
    return new_list

print(f(lambda e: e**2, range(4)))          # [0, 1, 4, 9, 16, 25]
print(f(lambda e: e**2, range(10), lambda e: e % 2 == 0))  # [0, 4, 16]

````

````python
import random

def spin_slot_machine():
    """Simulates a single spin of a slot machine."""
    reels = ["Cherry", "Bell", "Lemon", "Orange", "Star", "Skull"]
    result = [random.choice(reels), random.choice(reels), random.choice(reels)]
    return result

def check_win(result):
    """Checks if the spin result is a win."""
    if result[0] == result[1] == result[2]:
        print("JACKPOT! Three of a kind!")
        return True
    elif result[0] == result[1] or result[1] == result[2] or result[0] == result[2]:
        print("Two of a kind!")
        return True
    else:
        print("No win this spin.")
        return False

# Main game loop
print("Welcome to the simple slot machine!")
while input("Press Enter to spin (or type 'quit' to exit): ").lower() != 'quit':
    spin_result = spin_slot_machine()
    print("Spinning...")
    print(f"Result: {spin_result}")
    check_win(spin_result)

print("Thanks for playing!")
````

# Task
Create a simple dice game where the user guesses the sum of two dice rolls. The game should include a loop to allow multiple rounds.

````python
import random

def roll_dice():
    """Simulates rolling two dice and returns their sum."""
    die1 = random.randint(1, 6)
    die2 = random.randint(1, 6)
    return die1 + die2

def play_game():
    """Plays a round of the dice guessing game."""
    dice_sum = roll_dice()
    while True:
        try:
            user_guess = int(input("Guess the sum of two dice (2-12): "))
            if 2 <= user_guess <= 12:
                break
            else:
                print("Please enter a number between 2 and 12.")
        except ValueError:
            print("Invalid input. Please enter a number.")

    print(f"The dice rolled a total of {dice_sum}.")

    if user_guess == dice_sum:
        print("Congratulations! You guessed correctly!")
    else:
        print("Sorry, that's incorrect.")

# Game loop
print("Welcome to the Dice Guessing Game!")
while True:
    play_game()
    play_again = input("Play again? (yes/no): ").lower()
    if play_again != 'yes':
        break

print("Thanks for playing!")
````

````python
import random

def spin_slot_machine():
    """Simulates a single spin of a slot machine."""
    reels = ["Cherry", "Bell", "Lemon", "Orange", "Star", "Skull"]
    return [random.choice(reels) for _ in range(3)]

def check_win(result, bet):
    """Checks if the spin result is a win and returns the payout."""
    if result[0] == result[1] == result[2]:
        print("🎉 JACKPOT! Three of a kind!")
        return bet * 10   # jackpot = 10x payout
    elif result[0] == result[1] or result[1] == result[2] or result[0] == result[2]:
        print("✨ Two of a kind!")
        return bet * 2    # 2x payout
    else:
        print("💀 No win this spin.")
        return 0

# --- Main game loop ---
credits = 20
print("🎰 Welcome to the slot machine!")
print(f"You start with {credits} credits.")

while credits > 0:
    print(f"\nYou have {credits} credits.")
    bet_input = input("Place your bet (or type 'quit' to cash out): ").lower()
    if bet_input == "quit":
        break

    # validate bet
    if not bet_input.isdigit():
        print("❌ Please enter a valid number.")
        continue
    bet = int(bet_input)

    if bet <= 0:
        print("❌ Bet must be at least 1 credit.")
        continue
    if bet > credits:
        print("❌ You don't have enough credits for that bet.")
        continue

    # spin the reels
    credits -= bet
    print("\nSpinning...")
    spin_result = spin_slot_machine()
    print(f"Result: {spin_result}")

    # check winnings
    winnings = check_win(spin_result, bet)
    credits += winnings
    print(f"💰 You won {winnings} credits!")

print(f"\n🏁 Game over. You leave with {credits} credits.")
print("Thanks for playing! 🎲")
#🔹 Example Run
#sql
#Copy code
#🎰 Welcome to the slot machine!
#You start with 20 credits.

#You have 20 credits.
#Place your bet (or type 'quit' to cash out): 5

#Spinning...
#Result: ['Cherry', 'Cherry', 'Star']
#✨ Two of a kind!
#💰 You won 10 credits!

#You have 25 credits.
#Place your bet (or type 'quit' to cash out): quit

#🏁 Game over. You leave with 25 credits.
#Thanks for playing! 🎲
````

````python
def sum_digits(s):
  total = 0
  for char in s:
    if char in "0123456789":
        val = int(char)
        total += val
  return total

print(sum_digits("9836"))

````
